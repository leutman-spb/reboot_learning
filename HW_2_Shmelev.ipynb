Создать представление с полями:

Название колонки | Источник иформации
--- | ---
Номер заказа | столбец orderNumber таблицы orders
Название продукта | столбец productName таблицы products

Отобразить результат вызова представления

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем), отобразить.
Учитывайте, что заказ должен быть доставлен! Shipped = True

```
df = pd.read_sql ("""
create or replace view ords as
    select distinct
    orders.orderNumber as"Номер заказа",
    prod.productName as "Название продукта",
    prod.productCode as "Код продукта",
    prod.quantityOrdered as "Количество заказов",
    prod.priceEach as "Цена за один",
    prod.buyPrice as "Цена покупки",
    (prod.quantityOrdered*prod.priceEach) as "Итоговая сумма",
    ((prod.priceEach-prod.buyPrice)*prod.quantityOrdered) as "Прибыль",
    customers.customerName as "Заказчик",
    concat (customers.contactLastName, ' ', customers.contactFirstName) as "Имя представителя",
    customers.addressLine1 as "адрес"
    from orders as orders 
    inner join (select distinct
                products.productName,
                products.productCode,
                products.buyPrice,
                orderdetails.orderNumber,
                orderdetails.quantityOrdered,
                orderdetails.priceEach
                from products as products
                inner join orderdetails as orderdetails on products.productCode=orderdetails.productCode
                ) as prod on orders.orderNumber=prod.orderNumber
    inner join customers as customers on orders.customerNumber=customers.customerNumber
    order by orders.orderNumber, prod.productCode
""", connection)
```

```
# отправим запрос на подтверждение результата
df_view = pd.read_sql('''
SELECT * from ords''', connection)
df_view.head(10)
```

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем)

```
df_group = pd.read_sql('''
SELECT 
Заказчик,
sum(Прибыль) as "Суммарная прибыль" 
from ords
group by Заказчик
order by sum(Прибыль) desc
''', connection)
df_group.head(10)
```
